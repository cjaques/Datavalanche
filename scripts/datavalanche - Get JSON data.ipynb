```python
import numpy as np
import unicodedata
import requests
import json
from bs4 import BeautifulSoup
```

```python
import csv
rows = []
for i in range(2007, 2018):
    csv_reader = csv.reader(open('../data/data_avalanche/%d.csv' % i, 'rU'), delimiter=',')

    line_count = 0
    for row in csv_reader:
        if len(row) and 'http://www.data-avalanche.org/avalanche/' in row[0]:
            rows += [row[0].split('/')[4].split('"')[0]]
```

```python
def extract_data_dict_from_html(html):
    idx_start = html.find_all('script')[1].get_text().find('parse(\'')
    idx_end = html.find_all('script')[1].get_text().find("cache")
    s = (html.find_all('script')[1].get_text()[idx_start+7:idx_end-3] + '}')
    return json.loads(
        unicodedata.normalize('NFKD',s.encode("utf-8").decode("unicode_escape")).encode('ASCII', 'ignore'))
```

```python
def get_data_dict(idx='1528810866223'):
    url = 'http://www.data-avalanche.org/avalanche/' + idx
    r = requests.get(url)
    return extract_data_dict_from_html(BeautifulSoup(r.content, 'html.parser'))
```

```python
data = []
```

```python
for row in rows:
    try:
        data += [get_data_dict(str(row))]
        print "Success", row
    except:
        print "Failed", row
```

```python
np.save('avalanche-data', data)
```
